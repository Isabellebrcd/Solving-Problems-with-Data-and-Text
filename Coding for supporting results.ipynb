* how many times, on mean average, is each token used in the statement text

Store the results in a new column named `mean token use` in the unit_df data frame, to store how many times on average tokens are used in that university's statement text

```python
# please add your code for Q2 here (double click to edit)

# we calculate how many times, on mean average, is each token used in the text
#total number of tokens / number of unique tokens 
def calculate_mean_token_use(token_list):
    total = len(token_list)
    unique = len(set(token_list))
    return total / unique

unit_df['mean token use'] = unit_df['tokens'].apply(calculate_mean_token_use)
```

A function is created to calculate the mean token use. The function takes a list of token as argument. The total number of tokens in the list is stored in a 'total' variable. By using ``set``, the list is now without any repetition, so there is now only unique values. Then, the number of distinct tokens is stored in an 'unique' variable. The function returns the division of 'total' by 'unique' which is the mean average per unique token. The function is applied to each list of tokens in the 'tokens' column and the result is stored in a new column named 'mean token use'.

## More pre-processing

```python
#importing libraries
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
import string

lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(pos_tag):
    tag = pos_tag[0].upper()
    tag_dict = {"J": "a",  # Adjective
                "N": "n",  # Noun
                "V": "v",  # Verb
                "R": "r"}  # Adverb
    return tag_dict.get(tag, "n")

#creation of a function with more pre-processing things
def preprocess_text(text):
    # tokenisation
    tokens = word_tokenize(text)
    
    # lowercase and removing punctuation
    tokens = [token.lower() for token in tokens if token not in string.punctuation]
    
    #removing stopwords
    tokens = remove_stopwords(tokens)
    
    #lemmatisation (with POS tagging to be more precise)
    pos_tags = pos_tag(tokens)
    cleaned_tokens = [lemmatizer.lemmatize(token, get_wordnet_pos(pos_tag)) 
                      for token, pos_tag in pos_tags]
    
    return cleaned_tokens


unit_df['cleaned_tokens'] = unit_df['statement text'].apply(preprocess_text)
```

```python
unit_df[['statement text', 'cleaned_tokens']].head()
```

```python
#function to get all tokens from groups
def get_all_tokens(group):
    all_tokens = []
    for tokens in unit_df[unit_df['sample'] == group]['cleaned_tokens']:
        all_tokens.extend(tokens)
    return all_tokens

# getting tokenss for b_env and w_env
b_env_tokens = get_all_tokens('b_env')
w_env_tokens = get_all_tokens('w_env')

#calculate frequency distribution
b_env_freq = FreqDist(b_env_tokens)
w_env_freq = FreqDist(w_env_tokens)

#We want to print the 10 most commons tokens
print("most common tokens of b_env :")
print(b_env_freq.most_common(10))
print("\nmost common tokens of w_env :")
print(w_env_freq.most_common(10))

#mean token use calculation with previous function
b_env_mean_token_use = calculate_mean_token_use(b_env_tokens)
w_env_mean_token_use = calculate_mean_token_use(w_env_tokens)

#printing it for the two groups
print("\nMean token use (total tokens / unique tokens) :")
print(f"b_env : {b_env_mean_token_use:.2f}")
print(f"w_env : {w_env_mean_token_use:.2f}")
```

## Topic modelling

```python
from gensim import corpora

#creation of Dict and corpus 'bag of words' for LDA
def prepare_lda_data(group):
    group_sentences = unit_df[unit_df['sample'] == group]['cleaned_tokens'].tolist()
    dictionary = corpora.Dictionary(group_sentences)
    corpus = [dictionary.doc2bow(tokens) for tokens in group_sentences]
    return dictionary, corpus, group_sentences

#apply function to the two groups
b_env_dict, b_env_corpus, b_env_sentences = prepare_lda_data('b_env')
w_env_dict, w_env_corpus, w_env_sentences = prepare_lda_data('w_env')

print(f"Number of doc in b_env : {len(b_env_sentences)}")
print(f"Number of doc in w_env : {len(w_env_sentences)}")
```

```python
from gensim.models import LdaModel

#function to train LDA model with 5 topics and 10 passes
def train_lda_model(corpus, dictionary, num_topics=5, passes=10):
    lda_model = LdaModel(
        corpus=corpus,
        id2word=dictionary,
        num_topics=num_topics,
        passes=passes,
        random_state=42,
        alpha='auto',
        per_word_topics=True
    )
    return lda_model

#LDA for b_env
b_env_lda_5topics_10passes = train_lda_model(b_env_corpus, b_env_dict, num_topics=5, passes=10)
#LDA for w_env
w_env_lda_5topics_10passes = train_lda_model(w_env_corpus, w_env_dict, num_topics=5, passes=10)

#print topics
print("Topics for b_env (5 topics, 10 passes) :")
for idx, topic in b_env_lda_5topics_10passes.print_topics(num_words=10):
    print(f"Topic #{idx}: {topic}")

print("\nTopics for w_env (5 topics, 10 passes) :")
for idx, topic in w_env_lda_5topics_10passes.print_topics(num_words=10):
    print(f"Topic #{idx}: {topic}")
```

## Sentiment Analysis

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from transformers import pipeline

#vader
sid = SentimentIntensityAnalyzer()

#Roberta
roberta = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment")
```

```python
def vader_scores(text):
    scores = sid.polarity_scores(text)
    return scores['compound']

#vader on statement text
unit_df['vader_score'] = unit_df['statement text'].apply(vader_scores)
```

```python
def roberta_scores(text):
    #roberta limited at 512 characters
    text = text[:512]
    result = roberta(text)
    scores = {
        'neg': 0.0,
        'neu': 0.0,
        'pos': 0.0
    }
    label = result[0]['label']
    score = result[0]['score']
    if label == 'LABEL_0':
        scores['neg'] = score
    elif label == 'LABEL_1':
        scores['neu'] = score
    elif label == 'LABEL_2':
        scores['pos'] = score
    #score : (pos - neg), between [-1 1]
    global_score = (scores['pos'] - scores['neg'])
    return global_score, scores

#apply roberta function
unit_df[['roberta_score', 'roberta_scores']] = unit_df['statement text'].apply(roberta_scores).apply(pd.Series)
```

```python
def categorize_score(score):
    if score >= 0.05:
        return 'pos'
    elif score <= -0.05:
        return 'neg'
    else:
        return 'neu'

unit_df['vader_sentiment'] = unit_df['vader_score'].apply(categorize_score)
unit_df['roberta_sentiment'] = unit_df['roberta_score'].apply(categorize_score)

vader_dist = unit_df.groupby(['sample', 'vader_sentiment']).size().unstack(fill_value=0)
roberta_dist = unit_df.groupby(['sample', 'roberta_sentiment']).size().unstack(fill_value=0)

print("\nVader sentiment distribution:")
print(vader_dist)
print("\nRoberta sentiment distribution:")
print(roberta_dist)
```

## Relationships

```python
import seaborn as sns
from scipy.stats import linregress

#total doctoral degrees awarded and total research income by universities
docAwards['total_doctoral'] = docAwards[year_columns].sum(axis=1)
total_doctoral_uni = docAwards.groupby(['Institution name', 'Sample'])['total_doctoral'].sum().reset_index()
total_income_uni = researchIncome.groupby('Institution name')['Total income for academic years 2013-14 to 2019-20'].sum().reset_index()

merged_data = total_doctoral_uni.merge(
    total_income_uni,
    on='Institution name'
)

plt.figure(figsize=(10, 6))
palette = dict(zip(['ENV BETTER THAN KENT', 'ENV WORSE THAN KENT'], colors))
sns.scatterplot(x='total_doctoral', y='Total income for academic years 2013-14 to 2019-20', hue='Sample', s= 100, palette=palette, data=merged_data)

slope, intercept, r_value, p_value, std_err = linregress(merged_data['total_doctoral'], merged_data['Total income for academic years 2013-14 to 2019-20'])
x = np.array(merged_data['total_doctoral'])
y = slope * x + intercept
plt.plot(x, y, color='black', linestyle='--')

plt.title("Total doctoral degrees awarded vs total research Income")
plt.xlabel("Total doctoral degrees awarded from 2°13 to 2019")
plt.ylabel("total research Income (£)")
plt.legend()
plt.show()

#calculation of correlation
correlation = merged_data['total_doctoral'].corr(merged_data['Total income for academic years 2013-14 to 2019-20'])
print("\ncorrelation Total doctoral degrees awarded vs total research Income : ",correlation)
```

```python
ov_fte = resultsSummary[resultsSummary['Profile'] == 'Overall']
total_fte_uni = ov_fte.groupby(['Institution name', 'Sample'])['FTE of submitted staff'].sum().reset_index()

merged_data2 = total_doctoral_uni.merge(
    total_fte_uni,
    on=['Institution name', 'Sample']
)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='FTE of submitted staff', y='total_doctoral', hue='Sample', 
                palette=palette, s=100, data=merged_data2)

slope, intercept, r_value, p_value, std_err = linregress(merged_data2['FTE of submitted staff'], merged_data2['total_doctoral'])
x = np.array(merged_data2['FTE of submitted staff'])
y = slope * x + intercept
plt.plot(x, y, color='black', linestyle='--')

plt.title("FTE Submitted Staff vs Total doctoral degrees awarded")
plt.xlabel("FTE Submitted Staff")
plt.ylabel("Total doctoral degrees awarded")
plt.legend()
plt.show()

correlation = merged_data2['FTE of submitted staff'].corr(merged_data2['total_doctoral'])
print("\ncorrelation FTE submitted staff and Total doctoral degrees awardeds  :" ,correlation)
```

```python
# classification
uk_public_sources = [
    'BEIS Research Councils, The Royal Society, British Academy and The Royal Society of Edinburgh',
    'UK central government bodies/local authorities, health and hospital authorities',
    'UK central government tax credits for research and development expenditure'
]

uk_charities_sources = [
    'UK-based charities (open competitive process)',
    'UK-based charities (other)'
]

uk_industry_other_sources = [
    'UK industry, commerce and public corporations',
    'UK other sources'
]

eu_funding_sources = [
    'EU government bodies',
    'EU-based charities (open competitive process)',
    'EU industry, commerce and public corporations',
    'EU (excluding UK) other'
]

non_eu_funding_sources = [
    'Non-EU-based charities (open competitive process)',
    'Non-EU industry commerce and public corporations',  
    'Non-EU other'
]

#new coloumn for the classification
def assign_class(source):
    if source == 'Total income':
        return 'Exclude'  
    elif source in uk_public_sources:
        return 'UK Public Funding'
    elif source in uk_charities_sources:
        return 'UK Charities'
    elif source in uk_industry_other_sources:
        return 'UK Industry and Other'
    elif source in eu_funding_sources:
        return 'EU Funding'
    elif source in non_eu_funding_sources:
        return 'Non-EU Funding'
    else:
        return 'Other'

researchIncome['Category'] = researchIncome['Income source'].apply(assign_class)
researchIncome_filtered = researchIncome[researchIncome['Category'] != 'Exclude']
print(researchIncome_filtered['Category'].value_counts())
```

```python
#total income by category and the two group
income_cat = researchIncome_filtered.groupby(['Sample', 'Category'])['Total income for academic years 2013-14 to 2019-20'].sum().unstack(fill_value=0)
total_income = researchIncome[researchIncome['Income source'] == 'Total income'].groupby('Sample')['Total income for academic years 2013-14 to 2019-20'].sum()
income_cat_perc = income_cat.div(total_income, axis=0) * 100

print("\nTotal income for the two groups :")
print(total_income.round(0))
print("\nTotal income for the categories:")
print(income_cat.round(0))
print("\nTotal income in percentage for the categories and for the two groups:")
print(income_cat_perc.round(2))

fig, axes = plt.subplots(1, 2, figsize=(12, 6))
groups = ['ENV BETTER THAN KENT', 'ENV WORSE THAN KENT']
categories = income_cat_perc.columns

for i, group in enumerate(groups):
    axes[i].pie(income_cat_perc.loc[group], labels=categories, autopct='%1.1f%%', startangle=90)
    axes[i].set_title(f"Classification\n{group}")
plt.show()
```
